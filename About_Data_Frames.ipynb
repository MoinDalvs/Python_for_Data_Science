# Merge Data

```python
movies=pd.DataFrame({'movie_id':[1,2,3,4,5],
                    'movie_title':['Toy Story 2','Mission Impossible','Avatar','Blade Runner','Ghost Rider']})
movies
```

```python
members= DataFrame({'mem_id':[1,2,3,4],
                    'name':['Tom','John','Aron','Bobby'],
                    'movie_id':[1,2,3,nan]})
members
```

```python
pd.merge(movies,members, on='movie_id',how='left')
```

```python
pd.merge(members,movies,on='movie_id',how='right')
```

```python
pd.merge(members,movies,on='movie_id',how='left')
```

```python
pd.merge(movies,members,on='movie_id',how='outer')
```

```python
pd.merge(members,movies,on='movie_id',how='outer',indicator=True)
```

```python
pd.merge(movies,members,on='movie_id')
```

```python
pd.merge(members,movies,how='inner',on='movie_id')
```

```python
add_mem={'mem_id':5,'name':'Amy','movie_id':3}
members_updated=members.append(add_mem,ignore_index=True)
```

```python
members_updated
```

```python
pd.merge(members_updated,movies,on='movie_id',how='left')
```

```python
pd.merge(members_updated,movies,on='movie_id')[['movie_title','name']]
```
